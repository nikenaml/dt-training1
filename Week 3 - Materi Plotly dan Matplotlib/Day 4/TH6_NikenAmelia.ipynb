```python
# Lengkapi Library yang dibutuhkan
import matplotlib.pyplot as plt

# Mendefinsikan Data dalam bentuk list
data = {'pineapple': 15, 'grapes': 5, 'gauva': 40, 'watermelon': 20}
names = list(data.keys())
values = list(data.values())

# Lengkapi code untuk mendefinisikan visualisasi dengan bar, scatter dan plot
fig, axs = plt.subplots(1, 3, figsize=(15, 4), sharey=True)

# lengkapi code fungsi bar
axs[0].bar(names, values, color='y')

# lengkapi code fungsi scatter 
axs[1].scatter(names, values, color="m", marker="p", s=300)

# lengkapi code fungsi plot
axs[2].plot(names, values, color="r")


fig.suptitle('Categorical Plotting')

plt.ioff() 
```

```python
# Lengkapi Library yang dibutuhkan
import pandas as pd
import numpy as np

# Lengkapi dan periksa code Mendefinisikan Data
raw_data={'names':['Php','Python','R','Matlab','Java'],
         'jan':[123,124,125,126,127],
         'feb':[100,94,115,100,29],
         'march':[3,5,7,6,9]}

# lengkapilah code mendefiniskan data menjadi data frame 
df=pd.DataFrame(raw_data,columns=['names','jan','feb','march'])
df['total']=df['jan']+df['feb']+df['march']
df
```

```python
fig, ax = plt.subplots() 

# lengkapilah code Mendefiniskan fungsi visualisasi dengan menggunakan pie chart
ax.pie(df['total'],labels=df['names'],autopct='%1.1f%%')
ax.axis('equal')
plt.show()
```

```python
# lengkapi code fungsi bar 
fig, ax = plt.subplots() 
ax.bar(df['names'],df['jan'], label='Januari', color="r")
ax.bar(df['names'],df['feb'], label='Februari', color="b")

ax.legend()
plt.xlabel('nama bahasa pemograman')
plt.ylabel('jumlah per bulan')

plt.title('perbandingan penggunaan bahasa pemograman perbulan')

plt.show()
```
