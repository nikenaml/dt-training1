```python
print("Expected Result")

# import Library pandas sebagai pd dan matplotlip sebagai plt
import pandas as pd
import matplotlib.pyplot as plt

# Read data tugas 
df_data = pd.read_csv("Tugas.csv")
print(df_data)

# lengkapi code Mendefiniskan Fungsi Figure()
fig = plt.figure()
ax = fig.add_subplot()

# Lengkapi sintax pada code untuk mendefinisikan Data 
data_x = df_data["nama"]
data_y = df_data["uts"]

# Lengkapi code menambahkan judul pada sumbu x dan sumbu y
plt.title("line chart Nilai UTS Siswa")
plt.figure("Nama siswa")
plt.figure("Nilai Uts")

# Lengkapi code menambahkan objek pada axex
ax.plot(data_x,data_y, color="red", marker="s")


# menampilkan data
print("-------------- Hasil Visualisasi-------------")
plt.show()
```
